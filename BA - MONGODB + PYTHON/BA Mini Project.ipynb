## Overview of the project -
This project involves the exploration and analysis of a books dataset obtained from Kaggle. The dataset contains various attributes of books, including titles, authors, categories, prices, ratings, and more. The primary goal is to leverage MongoDB Atlas, a cloud-based database service, to store this dataset and perform data operations using Python and PyMongo.

## Project Steps:

Data Collection: Initially, a dataset consisting of book-related JSON data was acquired from Kaggle.

Database Setup: Using MongoDB Atlas, a cloud-based database service, a database named "Library" was created. Within this database, a collection named "books" was established to store the book-related information.

Data Import: The collected JSON data was inserted into the MongoDB Atlas collection "books" to make it accessible for querying and analysis.

Python Connection: The MongoDB database hosted on MongoDB Atlas was connected to a Python environment using the provided URI through the PyMongo library.

Data Operations: Utilizing PyMongo's functionalities, various data operations were performed on the book dataset stored in MongoDB. #### Grouping Operations

1) Group and count by category

```python
# Grouping operation by 'category' field
grouped_data_category = books.aggregate([
    {
        "$group": {
            "_id": "$category",  # Group by the 'category' field
            "count": {"$sum": 1}  # Count the number of documents in each group
            # Additional aggregation operations can be performed here
        }
    }
])

# Display the grouped data
for category in grouped_data_category:
    print(category)
```

2) Group and count by ratings

```python
# Grouping operation by 'stars' field
grouped_by_rating = books.aggregate([
    {
        "$group": {
            "_id": "$stars",  # Group by the 'stars' field
            "count": {"$sum": 1}  # Store the documents in an array per rating
        }
    }
])

# Display the grouped data
for rating_group in grouped_by_rating:
        pprint(rating_group)
```

#### Sorting Operations

1) Sorting in ascending order of price

```python
# Sort the data according to price in ascending order
sorted_by_price = list(books.find().sort("price", 1))
print("\nSorted by Price:")
for book in sorted_by_price:
    pprint(book)
```

2) Sorting in increasing order of availability

```python
# Sort the data according to availability in ascending order
sorted_by_availability = list(books.find().sort("availability", 1))
print("\nSorted by Availability:")
for book in sorted_by_availability:
    pprint(book)
```

3) Sorting in Descending order of stars

```python
# Sort the data according to ratings in descending order
sorted_by_ratings = list(books.find().sort("stars", -1))
# Print or use the sorted data as needed
print("Sorted by Ratings:")
for book in sorted_by_ratings:
    pprint(book)
```

#### Filter Operations

1) Filtering by price

```python
# Define the price range for the filter
min_price = 10
max_price = 20

# Filter operation on the 'price' field
filtered_books = books.find({"price": {"$gte": min_price, "$lte": max_price}})

# Display the filtered books
for book in filtered_books:
    pprint(book)
```

2) Filter by rating

```python
# Filter operation on the 'stars' field for books with a 5-star rating
filtered_books = books.find({"stars": 5})

# Display the books with a 5-star rating
for book in filtered_books:
    pprint(book)
```

3) Filter by category

```python
# Filter operation on the 'category' field for books categorized as 'thriller'
filtered_books = books.find({"category": "thriller"})

# Display the books categorized as 'thriller'
for book in filtered_books:
    pprint(book)
```
