### Loading Libraries and Data

### Transformation

```python
data.diagnosis=[1 if each=="M" else 0 for each in data.diagnosis]
y=data.diagnosis.values
x_data=data.drop(["diagnosis","id"],axis=1)
```

### Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x = scaler.fit_transform(x_data)
```

### Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

### Model building

#### 1) Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
model_nb = nb.fit(x_train, y_train)
y_pred = model_nb.predict(x_test)
report = classification_report(y_test, y_pred)
print(report)
```

#### 2) KNN

```python
from sklearn.neighbors import KNeighborsClassifier

# Initialize the KNN model
knn = KNeighborsClassifier(n_neighbors=5)  # You can change the number of neighbors (k) as needed

# Train the KNN model on the training data
model_knn = knn.fit(x_train, y_train)

# Make predictions on the test data
y_pred_knn = model_knn.predict(x_test)

# Calculate accuracy for the KNN model
accuracy_knn = accuracy_score(y_pred_knn, y_test)
print("K-Nearest Neighbors Accuracy:", accuracy_knn)

# Print the classification report for the KNN model
from sklearn.metrics import classification_report
report_knn = classification_report(y_test, y_pred_knn)
print("Classification Report for K-Nearest Neighbors:\n", report_knn)
```
